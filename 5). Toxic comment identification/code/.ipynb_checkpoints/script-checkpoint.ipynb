```python
#create new feature count of  toxic
def count_toxic(row):
    count=0
    for i in range(2,8):
        if row[i]==1:
            count+=1
    return count
    
    
train_df['toxic_level']=train_df.apply(count_toxic,axis=1)
```

```python
# sort the data based on count of toxic
sorted_comment_df=train_df.sort_values(by='toxic_level', ascending=False)
sorted_comment_df=sorted_comment_df.head(data_limiter)
sorted_comment_df.head()['comment_text']
```

```python
#remove extra words
def remove_extra_text( review, remove_stopwords=False ):
    # Function to convert a document to a sequence of words,
    # optionally removing stop words.  Returns a list of words.
    #
    # 1. Remove HTML
    review_text=BeautifulSoup(review).get_text()
    #
    # 2. Remove non-letters
    review_text = re.sub("[^a-zA-Z]"," ", review_text)
    #
    # 3. Convert words to lower case and split them
    words = review_text.lower().split()
    #
    # 4. Optionally remove stop words (false by default)
    if remove_stopwords:
        stops = set(stopwords.words("english"))
        words = [w for w in words if not w in stops]

    #
    # 5. Return a list of words
    return(' '.join(words))


```

```python
#actual cleaning 
def clean_text(row):
    return remove_extra_text(row['comment_text'])
sorted_comment_df['comment_text']=sorted_comment_df.apply(clean_text,axis=1)
```

```python
#sorted and clean data
sorted_comment_df.head(10)
```

```python
#text to vector 
vectorizer = CountVectorizer(analyzer = "word")
converted_comment_to_vector=vectorizer.fit_transform(sorted_comment_df['comment_text'])
```
